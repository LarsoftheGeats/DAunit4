```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("world_population.csv")
```

```python
df.head()
```

```python
df.drop(['CCA3', 'Capital','Continent','Area (km²)','Density (per km²)','World Population Percentage'], axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(['Growth Rate'], axis=1, inplace=True)
```

```python
df=pd.melt(df, id_vars =['Country/Territory'], value_vars =['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population'])
```

```python
df=df.rename(columns={'variable':'year'})
```

```python
df['year']=df.year.str.replace(" Population","")
```
